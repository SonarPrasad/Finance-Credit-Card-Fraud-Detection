```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/creditcard.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
# Plot the class distribution
plt.figure(figsize=(12, 8))
sns.countplot(data['Class'])
plt.title("Class Distribution")
plt.xlabel("Class (0: Legitimate, 1: Fraud)")
plt.ylabel("Count")
plt.show()

# Print the percentage of fraud cases
fraud_percentage = data['Class'].value_counts(normalize=True) * 100
print("Fraud Percentage:", fraud_percentage[1], "%")
```

### Class Distribution Plot
This plot shows the distribution of classes in the dataset, where Class 0 represents legitimate transactions and Class 1 represents fraudulent transactions. The overwhelming height of the bar for legitimate transactions indicates a highly imbalanced dataset, which is common in fraud detection scenarios. This imbalance can pose challenges for model training and evaluation.

```python
# Plot time distributions
plt.figure(figsize=(12, 8))
sns.histplot(data[data['Class'] == 0]['Time'], bins=50, color='blue', label='Legitimate', kde=True)
sns.histplot(data[data['Class'] == 1]['Time'], bins=50, color='red', label='Fraudulent', kde=True)
plt.legend()
plt.title("Transaction Time Distribution")
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency")
plt.show()
```

### Transaction Time Distribution Plot
This plot illustrates the distribution of transaction times for legitimate and fraudulent transactions. The blue bars represent legitimate transactions, while the red line indicates fraudulent transactions. The plot reveals that legitimate transactions occur more frequently at certain times, while fraudulent transactions are less common and may cluster around specific time intervals.

```python
# Plot transaction amount distributions
plt.figure(figsize=(12, 8))
sns.boxplot(x='Class', y='Amount', data=data)
plt.title("Transaction Amount by Class")
plt.xlabel("Class (0: Legitimate, 1: Fraud)")
plt.ylabel("Transaction Amount")
plt.show()
```

### Transaction Amount by Class Plot
This box plot displays the distribution of transaction amounts for legitimate and fraudulent transactions. The box for legitimate transactions shows a wide range of amounts with several outliers, indicating significant variability. In contrast, the box for fraudulent transactions is narrower, suggesting that they tend to have lower amounts. This visualization highlights the financial characteristics of transactions and can assist in detecting fraudulent activities based on transaction amounts.

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=np.log(data['Amount'] + 1), bins=50, kde=True, color='blue')  # Logarithmic transformation
plt.title("Distribution of Transaction Amounts (Logarithmic)")
plt.xlabel("Log Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

### Insights from the plot:

- **Skewed Distribution:** The plot confirms that the transaction amounts are highly skewed to the right. This means that there are many small transactions and a few very large ones.
- **Logarithmic Scale:** The logarithmic scale on the x-axis helps to better visualize the distribution of the smaller transactions. It compresses the larger values and spreads out the smaller ones, making the distribution appear more evenly distributed.
- **Mode:** The peak of the histogram indicates the most frequent transaction amount. In this case, it seems to be around 1 on the log scale, which corresponds to a transaction amount of approximately $10.
- **Outliers:** There are a few outliers on the right side of the plot, representing very large transactions. These outliers can significantly impact the mean and standard deviation of the data.

```python
# Convert seconds to hours
data['Hours'] = data['Time'] / 3600

# Group by hour and count transactions
transactions_per_hour = data.groupby('Hours').size()

# Create the line plot
plt.figure(figsize=(12, 8))
plt.plot(transactions_per_hour.index, transactions_per_hour.values, marker='o')
plt.title("Transactions Over Time")
plt.xlabel("Hours Since First Transaction")
plt.ylabel("Number of Transactions")
plt.grid()
plt.show()
```

### Insights from the Plot

- **Non-Uniform Transaction Distribution**: The plot shows that transactions are not uniformly distributed over time. Instead, there are clusters of high transaction activity interspersed with periods of lower activity.

- **Peak Periods**: There are several distinct peaks in the plot, indicating time periods with significantly higher transaction volumes. These peaks might correspond to specific events, promotions, or daily/weekly patterns in consumer behavior.

- **Variability**: The plot highlights the variability in transaction rates over time. This could be due to various factors such as seasonal trends, day-of-week effects, or specific marketing campaigns.

```python
# Select a subset of features for pair plotting
features_to_plot = ['V1', 'V2', 'V3', 'V4', 'Amount', 'Class']
plt.figure(figsize=(12, 8))
sns.pairplot(data[features_to_plot], hue='Class', diag_kind='kde', palette={0: "blue", 1: "red"})
plt.suptitle("Pair Plot of Selected Features", y=1.02)
plt.show()
```

### Insights from the Pair Plot

**Feature Distribution**:
- The diagonal plots show the distribution of individual features. We can observe the shape, spread, and outliers for each feature.
- Some features (like V1, V2, V3, and Amount) seem to have a normal-like distribution, while others might have a skewed distribution.

**Relationships Between Features**:
- The off-diagonal plots show the relationship between pairs of features.
- Some features seem to have a linear relationship, while others might have a more complex relationship.
- There are some interesting patterns in the data, such as clusters and outliers.

**Class Separation**:
- The different colors represent different classes (likely '0' and '1').
- The overlap between the colors suggests that it might be challenging to separate the classes based on these features alone. Some features, like V1, V2, and V3, show some separation between the classes, but it's not very distinct.

```python
plt.figure(figsize=(12, 8))

# Violin plot for transaction amount distribution by class
sns.violinplot(x='Class', y='Amount', data=data, palette={'0': "blue", '1': "red"}, scale='width')

# Overlay mean values for each class
means = data.groupby('Class')['Amount'].mean()
for i, mean in enumerate(means):
    plt.scatter(x=i, y=mean, color='black', marker='o', label=f"Mean: Class {i}")

# Add plot details
plt.title("Distribution of Transaction Amounts by Class")
plt.xlabel("Class (0: Legitimate, 1: Fraud)")
plt.ylabel("Transaction Amount")
plt.legend(loc='upper right')
plt.show()
```

### Insights from the Distribution of Transaction Amounts by Class

- **Distribution Overview**:
  - The violin plot illustrates the distribution of transaction amounts for legitimate (Class 0) and fraudulent (Class 1) transactions.
  - The plot combines the features of a box plot and a density plot, providing a comprehensive view of the data distribution.

- **Class 0 (Legitimate Transactions)**:
  - The distribution for legitimate transactions shows a wide range of amounts, with a significant number of outliers on the higher end.
  - The density is concentrated around lower transaction amounts, indicating that most legitimate transactions are relatively small.

- **Class 1 (Fraudulent Transactions)**:
  - The distribution for fraudulent transactions is much narrower, suggesting that fraudulent transactions tend to have lower amounts.
  - The density is concentrated at the lower end, with very few transactions exceeding a certain threshold.

- **Mean Values**:
  - The plot includes markers for the mean transaction amounts for each class, highlighting that the mean for Class 0 is significantly higher than that for Class 1.
  - This difference in means reinforces the observation that fraudulent transactions are generally of lower value.

- **Implications for Fraud Detection**:
  - The stark contrast in transaction amounts between the two classes can be useful for developing fraud detection models.
  - Understanding these financial characteristics can assist in identifying potentially fraudulent activities based on transaction amounts.

```python
# Bin time into intervals (e.g., hours) and count occurrences
data['TimeBin'] = pd.cut(data['Time'], bins=np.arange(0, data['Time'].max(), 3600))
heatmap_data = data.groupby(['TimeBin', 'Class']).size().unstack(fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data.T, cmap='coolwarm', annot=False, cbar=True)
plt.title("Heatmap of Transactions Over Time by Class")
plt.xlabel("Time (Binned by Hour)")
plt.ylabel("Class (0: Legitimate, 1: Fraud)")
plt.xticks(rotation=45)
plt.show()
```

### Insights from the Heatmap of Transactions Over Time by Class

- **Overview**: This heatmap visualizes the frequency of transactions over time, categorized by class (0 for legitimate transactions and 1 for fraudulent transactions). The color intensity indicates the number of transactions, with darker colors representing higher frequencies.

- **Class Distribution**:
  - The bottom row (Class 1) shows a predominantly blue color, indicating a significantly lower number of fraudulent transactions compared to legitimate ones (Class 0).
  - The top row (Class 0) displays varying shades of color, suggesting that legitimate transactions occur more frequently across different time intervals.

- **Temporal Patterns**:
  - The heatmap reveals distinct patterns in transaction activity over time. There are periods where legitimate transactions peak, while fraudulent transactions remain sparse.
  - The presence of red and orange shades in the legitimate class indicates specific time intervals with notably high transaction volumes.

- **Fraud Detection Implications**:
  - The stark contrast in transaction frequencies between the two classes suggests that fraudulent transactions are rare and may cluster around specific times.
  - Understanding these temporal patterns can assist in developing strategies for real-time fraud detection, allowing for targeted monitoring during peak transaction periods.

```python
# Correlation matrix
corr = data.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap='coolwarm', annot=False, fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

### Feature Correlation Heatmap
Weak Correlations: Most features exhibit weak correlations with each other, indicating a low degree of linear dependence. This suggests that the features are relatively independent.

### Conclusion of Exploratory Data Analysis (EDA)

The Exploratory Data Analysis (EDA) on the credit card transaction dataset has provided us with important insights into the patterns and characteristics of both legitimate and fraudulent transactions.

#### Key Findings from Our Analysis:

- **Class Imbalance**:
  The dataset is highly imbalanced, with a significantly larger number of legitimate transactions (Class 0) compared to fraudulent ones (Class 1). This imbalance was clearly visible in the class distribution plot, where the majority of the transactions are legitimate. Such a disparity can lead to challenges in model training and evaluation, as models may become biased towards predicting the majority class. Addressing this imbalance will be crucial in building an effective fraud detection system.

- **Transaction Time Patterns**:
  When analyzing the transaction times, we noticed distinct patterns: legitimate transactions occurred more frequently at certain times, while fraudulent transactions were sparse and tended to cluster around specific intervals. These temporal patterns could offer valuable insights into identifying suspicious activities, suggesting that time-based features may play a significant role in detecting fraud.

- **Transaction Amount Analysis**:
  The distribution of transaction amounts revealed interesting differences: legitimate transactions spanned a wide range, with several high-value outliers, whereas fraudulent transactions mostly involved lower amounts. This suggests that the transaction amount could be a useful feature in identifying potential fraud, as fraudulent transactions may be characterized by smaller amounts.

- **Feature Relationships**:
  The pair plot revealed some linear correlations and potential clusters among the selected features. However, the overlap between the two classes indicated that distinguishing between legitimate and fraudulent transactions based solely on these features might be challenging. This finding highlights the need for more advanced techniques or additional features to improve classification accuracy.

- **Heatmap Insights**:
  The heatmap showing the frequency of transactions by class over time further emphasized the stark contrast between legitimate and fraudulent transactions. The scarcity of fraudulent transactions, particularly during peak times for legitimate transactions, highlights the importance of targeted monitoring to detect fraud in real time.
